```python
# Loads the pandas library to this program
import pandas as pd

# Reads the contents of excel file "board2.xlsx" 
df = pd.read_excel("board2.xlsx")
```

```python
# Creates a dataframe with columns "Name", "GEAS", "Electronics"
# Outputs the datas that satisfy the conditions given: (1) track MUST be instrumentation; (2) hometown MUST be Luzon; (3) Grades in Electronics > 70

Instru = df[
        (df["Track"] == "Instrumentation") & 
        (df["Hometown"] == "Luzon") & 
        (df["Electronics"] > 70)
        ][["Name", "GEAS", "Electronics"]]

Instru
```

```python
# Creates a new column by getting the mean of the three components: Math, Electronics, & GEAS.
df["Average"] = df[["Math", "Electronics", "GEAS"]].mean(axis=1)

# Creates a dataframe with columns "Name", "Track", "Electronics", & "Average"
# Outputs the data that satisfy the conditions given: (1) Must be from Mindanao; (2) Must be a female; (3) Must have an average of at least 55

Mindy = df[
        (df["Hometown"] == "Mindanao") & 
        (df["Gender"] == "Female") & 
        (df["Average"] >= 55)
        ][["Name", "Track", "Electronics", "Average"]]

Mindy
```

```python
# Loads the matplotlib library to this program
import matplotlib.pyplot as plt

# Compute the average scores then groups it by Track, Gender, & Hometown

avg_track = df.groupby("Track")["Average"].mean()
avg_gender = df.groupby("Gender")["Average"].mean()
avg_hometown = df.groupby("Hometown")["Average"].mean()

# Creates subplots to show comparisons side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plots the mean scores by Track
avg_track.plot(kind="bar", ax=axes[0], color="lightskyblue", rot=0)
axes[0].set_title("Average Scores by Track")
axes[0].set_ylabel("Average Score")

# Plots the mean scores by Gender
avg_gender.plot(kind="bar", ax=axes[1], color="lightsalmon", rot=0)
axes[1].set_title("Average Scores by Gender")
axes[1].set_ylabel("Average Score")

# Plots the mean scores by Hometown
avg_hometown.plot(kind="bar", ax=axes[2], color="lightgreen", rot=0)
axes[2].set_title("Average Scores by Hometown")
axes[2].set_ylabel("Average Score")

# Adjusts the layout of the graphs so that it will not overlap with each other.
plt.tight_layout()
plt.show()
```

```python
print ("The results of the visualization suggest that the track chosen in college has a greater impact on the average score, whereas gender and hometown show comparatively less contribution.")
```
